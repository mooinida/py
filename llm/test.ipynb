```python
import requests

def crawl_naver_all_search():
    url = "https://map.naver.com/p/api/search/allSearch"
    params = {
        "query": "성신여대역 맛집",
        "type": "all",
        "searchCoord": "127.0165;37.5927",
        "boundary": ""
    }
    headers = {
        "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_7)",
        "Referer": "https://map.naver.com/",
    }

    res = requests.get(url, headers=headers, params=params)

    print(f"📡 status code: {res.status_code}")
    print(f"📄 content preview: {res.text[:300]}")  # 응답 앞부분 출력

    if res.status_code != 200:
        print("❌ 요청 실패")
        return

    try:
        data = res.json()
        places = data.get("result", {}).get("place", {}).get("list", [])
        for idx, place in enumerate(places, start=1):
            print(f"{idx}. {place.get('name')} - {place.get('address')}")
    except Exception as e:
        print("⚠️ JSON 파싱 실패:", e)

crawl_naver_all_search()
```
